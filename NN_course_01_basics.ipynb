# Introduction

### A. Types of Neurons

* ** Linear Neurons **
    
    * $y = b + \sum_ix_iw_i$
    
    
* ** Binary Threshold Neurons **

    * $z = \sum_ix_iw_i$
    * $y = \begin{cases} 1 & \text{ if } z\geq\theta \\ 0 & \text{ otherwise } \end{cases}$
    
    
* ** Rectified Linear Neurons **

    * $z = b + \sum_ix_iw_i$
    * $ y = \begin{cases} z & \text{ if } z\gt 0 \\ 0 & \text{ otherwise } \end{cases}$
    
    
* ** Sigmoid Neurons **

    * $z = b + \sum_ix_iw_i$
    * $ y = \frac{1}{1 + e^{-z}} $
    
    
* ** Stochastic Binary Neurons **

    * $z = b + \sum_ix_iw_i$
    * $ p(s=1) = \frac{1}{1 + e^{-z}} $

### B. Perceptron Learning

* **Algorithm**

    * If the output unit is *correct*, leave its weights alone.
    * If the output unit *incorrectly outputs a 0*, add the input vector to the weight vector.
    * If the output unit *incorrectly outputs a 1*, subtract the input vector from the weight vector.
    
* **Convergence**

    * The learning will cause the weight to converge to a *feasible region* (i.e. the space where weights give correct prediction), if such a region exists.
    
* **Limitation**

    * With wrong features, perceptrons can never learn and converge. Therefore *learning features* is more crucial. Note that this is true only when the number of features is limited, because with no upper bound on the number of features, we can engineer a perceptron that can fit anything, but that wouldn't be an informative way to learn and the perceptron won't generalize.
